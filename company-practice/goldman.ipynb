```python
## Two-sum
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        d = {}
        for i, val in enumerate(nums):
            diff = target - val
            if val in d:
                return [i, d[nums[i]]]
            d[diff] = i
            
            
```

```python
## 696. Count Binary Substrings
class Solution:
    def countBinarySubstrings(self, s: str) -> int:
        prev, curr = 0,1
        count = 0
        for i in range(len(s)-1):
            if s[i] == s[i+1]:
                curr+=1
            else:
                count+= min(prev, curr)
                prev = curr
                curr = 1
        count+= min(prev, curr)
        return count
        
```

```python
## 647.Â Palindromic Substrings
class Solution:
    def countSubstrings(self, s: str) -> int:
        
        ans = len(s)
        
        for i in range(len(s)):
            j=1
            while i-j >=0 and i+j< len(s) and s[i+j] == s[i-j]:
                ans+=1
                j+=1
        
        for i in range(len(s)):
            k=0
            while i-k >=0 and i+j+1< len(s) and s[i-k] == s[i+k+1]:
                ans+=1
                k+=1
        
        return ans
```

```python
## 387. First Unique Character in a String
class Solution:
    def firstUniqChar(self, s: str) -> int:
        d={}
        seen = set()
        for i, val in enumerate(s):
            if val not in seen:
                d[val] = i
                seen.add(val)
            elif val in d:
                del d[val]
                
        return min(d.values()) if d else -1
```

```python
## 443. String Compression
class Solution:
    def compress(self, chars: List[str]) -> int:
        count = 1
        for i in range(len(chars)-1,-1,-1):
            if i and chars[i]==chars[i-1]:
                count += 1
                chars.pop(i)
            else:
                if count>1:
                    for item in str(count)[::-1]:
                        chars.insert(i+1, item)
                    count = 1
```

```python
# 780 -- Reaching Points
class Solution:
    def reachingPoints(self, sx: int, sy: int, tx: int, ty: int) -> bool:
        while ty>sy and tx>sx:
            ty, tx = tx%ty, ty%tx

        return (sx == tx and ty>=sy and (ty-sy)%sx==0) or \
        (sy==ty and tx>=sx and (tx-sx)%sy==0)
```

```python
## 1041 -- Robot Bounded in Circle
class Solution:
    def isRobotBounded(self, instructions: str) -> bool:
        dirX, dirY = 0,1
        x,y = 0,0

        for i in instructions:
            if i =="G":
                x +=dirX
                y+= dirY
            elif i=="L":
                dirX, dirY = dirY, -1*dirX
            else:
                dirX, dirY = -1*dirY, dirX

        return (x,y) == (0,0) or (dirX, dirY) != (0,1)
```

```python
## 1010. Pairs of Songs With Total Durations Divisible by 60
class Solution:
    def numPairsDivisibleBy60(self, time: List[int]) -> int:
        d = [0]*60
        count = 0
        for i in time:
            other_half = (60 - (i%60)) %60
            count += d[other_half]
            d[i%60] +=1
        return count
```
