```python
class BST:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
```

```python
def add_to_Binary_Search_Tree(node, value_to_add):
    if (node.value == value_to_add):
        return 
    
    if (node.value > value_to_add):
        if(node.left):
            add_to_Binary_Search_Tree(node.left, value_to_add)
        else: 
            node.left = BST(value_to_add)
    
    if (node.value < value_to_add):
        if(node.right):
            add_to_Binary_Search_Tree(node.right, value_to_add)
        else: 
            node.right = BST(value_to_add)

the_list = [5, 4, 6, 7, 1, 3, 1, 1, 1, 5]
node = BST(the_list[0])

for i in the_list[1:]:
    add_to_Binary_Search_Tree(node, i)
```

```python
print("In-order Traversal of our tree")
def inorder(node):
    if (node == None):
        return []
    return inorder(node.left) + [node.value] + inorder(node.right)
print(inorder(node))
print(" ")
```

```python
print("Pre-order Traversal of our tree")
def preorder(node):
    if (node == None):
        return []
    return [node.value] + preorder(node.left) + preorder(node.right)
print(preorder(node))
print(" ")
```
